```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import pandas as pd
from scipy.stats import boxcox
```

```python
df = pd.read_csv("sample_data.csv")
```

```python
# 原始分布
df[['AIR_TIME']].hist(bins=100)
```

```python
# BoxCox变换后的分布
df[['AIR_TIME']].apply(lambda x: boxcox(x)[0]).hist(bins=100)
```
